```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
with open('lagrangian.json') as infile:
    d = json.load(infile) 
```

```python
print(d[0].keys())
```

```python
for i in range(len(d)):
    plt.plot(d[i]['lon'],d[i]['lat'])
```

```python
data_list = []
for i in range(len(d)):
    a = pd.DataFrame(d[i])
    a['time'] = a.index*a.dt
    a.dropna(inplace = True)
    data_list.append(a) 
```
